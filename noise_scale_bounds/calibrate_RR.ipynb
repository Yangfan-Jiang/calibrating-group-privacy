```python
from scipy.special import comb, perm
from scipy import integrate
import numpy as np

from autodp import rdp_bank, rdp_acct, dp_acct,privacy_calibrator

from decimal import *
```

```python
def rdp2dp(rdp, bad_event, alpha):
    """
    convert RDP to DP, ref (Proposition 12):
    Canonne, ClÃ©ment L., Gautam Kamath, and Thomas Steinke. The discrete gaussian for differential privacy. In NeurIPS, 2020.
    """
    eps_dp = Decimal(rdp) + Decimal(1.0)/Decimal(alpha-1) * ((Decimal(1.0)/Decimal(bad_event)).ln() + Decimal(alpha-1)*Decimal(1-1.0/alpha).ln() - Decimal(alpha).ln())
    return float(eps_dp)

def basic_rdp_rr(alpha, p):
    rr_sum = Decimal(p)**Decimal(alpha) * Decimal(1-p)**Decimal(1-alpha) + Decimal(1-p)**Decimal(alpha) * Decimal(p)**Decimal(1-alpha)
    rdp_rr = rr_sum.ln() / Decimal(alpha-1)
    return float(rdp_rr)

def naive_group_rdp_rr(alpha, p, q, m):
    rr_rdp = basic_rdp_rr(alpha, p)
    
    """
    Suppose m=2^c for c \in Z
    """
    alpha = alpha * m

    func_rr = lambda x: rdp_bank.RDP_randresponse({'p': p}, x)

    # declare the moment accountants
    acct1 = rdp_acct.anaRDPacct()

    acct1.compose_poisson_subsampled_mechanisms(func_rr, q, coeff=1)

    rdp = acct1.get_rdp([alpha])[0]
    
    c = np.log2(m)
    
    return min(float(3**c * rdp), rr_rdp)

def exact_rdp_upd_rr(alpha, p, q, m):
    rdp_rr = Decimal(basic_rdp_rr(alpha, p))
    
    sum_ = Decimal(1-q)**Decimal(m) + (Decimal(1.0)-Decimal(1-q)**Decimal(m)) * Decimal(np.e)**Decimal(Decimal(alpha-1)*Decimal(rdp_rr))
    
    rdp = sum_.ln() / Decimal(alpha-1)
    return float(rdp)

def exact_rdp_lbd_rr(alpha, p, q, m):
    t1 = Decimal(Decimal(1-q)**Decimal(m) * Decimal(2*p-1.0) + Decimal(1-p))**Decimal(alpha) / (Decimal(p)**Decimal(alpha-1))
    t2 = Decimal(Decimal(1-q)**Decimal(m) * Decimal(1.0-2*p) + Decimal(p))**Decimal(alpha) / (Decimal(1-p)**Decimal(alpha-1))

    sum_ = t1 + t2
    
    lbd1 = sum_.ln() / Decimal(alpha-1)

    return float(lbd1)
```

```python
def calibrating_rr_noise_lbd(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 0.499
    sigma_min = 0.001
    
    def binary_search(left, right):
        mid = (left + right) / 2
        
        lbd = t*exact_rdp_lbd_rr(alpha, 1-mid, q, m)
        ubd = t*exact_rdp_lbd_rr(alpha, 1-left, q, m)
        
        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*exact_rdp_lbd_rr(alpha, 1-sigma_max, q, m) > tau:
        sigma_max *= 100
        if sigma_max > 1e10:
            print("sigma_max >", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return 1-sigma_max

def calibrating_rr_noise_naive(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 0.499
    sigma_min = 0.001
    
    def binary_search(left, right):
        mid = (left + right) / 2
        
        lbd = t*naive_group_rdp_rr(alpha, 1-mid, q, m)
        ubd = t*naive_group_rdp_rr(alpha, 1-left, q, m)

        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*naive_group_rdp_rr(alpha, 1-sigma_max, q, m) > tau:
        sigma_max *= 100
        if sigma_max > 1e10:
            print("sigma_max >", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return 1-sigma_max


def calibrating_rr_noise_base(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 0.499
    sigma_min = 0.001
    
    def binary_search(left, right):
        mid = (left + right) / 2
        
        lbd = t*basic_rdp_rr(alpha, 1-mid)
        ubd = t*basic_rdp_rr(alpha, 1-left)
        
        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*basic_rdp_rr(alpha, 1-sigma_max) > tau:
        sigma_max *= 100
        if sigma_max > 1e10:
            print("sigma_max >", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return 1-sigma_max


def calibrating_rr_noise_ours(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 0.499
    sigma_min = 0.001
    
    def binary_search(left, right):
        mid = (left + right) / 2
        
        lbd = t*exact_rdp_upd_rr(alpha, 1-mid, q, m)
        ubd = t*exact_rdp_upd_rr(alpha, 1-left, q, m)
        
        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*exact_rdp_upd_rr(alpha, 1-sigma_max, q, m) > tau:
        sigma_max *= 100
        if sigma_max > 1e10:
            print("sigma_max >", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return 1-sigma_max
```

## Subsampled RR Varying $m$

```python
#alpha = 16
p = 0.6
q = 0.03

iters = 1

m_list = [16, 24, 32, 48, 64, 96, 128, 192, 256]

alpha = 4
tau = 1.0

delta = 1e-5

dp_eps = rdp2dp(tau,delta,alpha)
print( "alpha={},tau={},eps={},".format(alpha, tau, dp_eps) )

naive_list = []
upd_list = []
basic_rr = []
lbd_list = []

for m in m_list:
    naive = 1e10
    if abs(np.log2(m)-int(np.log2(m))) < 0.00001:
        naive1 = calibrating_rr_noise_naive(q, m, alpha=alpha, tau=tau, t=iters, err=1e-4)
        naive2 = calibrating_rr_noise_base(q, m, alpha=alpha, tau=tau, t=iters, err=1e-4)
        naive = min(naive1, naive2)
    upd = calibrating_rr_noise_ours(q, m, alpha=alpha, tau=tau, t=iters, err=1e-4)
    lbd = calibrating_rr_noise_lbd(q, m, alpha=alpha, tau=tau, t=iters, err=1e-4)
    print("m={}, naive={},upd={},lbd={}".format(m, naive, upd, lbd))
    print("")
    naive_list.append(naive)
    upd_list.append(upd)
    lbd_list.append(lbd)
```
