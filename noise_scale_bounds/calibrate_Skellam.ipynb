```python
from scipy.special import comb, perm, gamma
from scipy import integrate
from scipy.stats import norm, laplace, skellam
import numpy as np

from autodp import rdp_bank, rdp_acct, dp_acct,privacy_calibrator

from decimal import *

getcontext().prec = 1024
```

```python
def get_sk_prob(lambda_, k):
    return Decimal(skellam.pmf(k, lambda_, lambda_))


def rdp_sk(alpha, lambda_, c):
    mu = Decimal(2.0*lambda_)
    A = ( Decimal(2*alpha-1) * Decimal(c)**Decimal(2.0) + Decimal(6*c) ) / (Decimal(4.0)*mu**Decimal(2.0))
    B = Decimal(3*c) / (Decimal(2)*mu)
    eps = Decimal(alpha)*Decimal(c)**Decimal(2.0) / (Decimal(2.0)*mu) + min(A, B)
    return eps


def rdp_sk_group(alpha, lambda_, c, k):
    mu = Decimal(2.0*lambda_)
    A = ( Decimal(2*alpha-1) * Decimal(k)**Decimal(2.0) * Decimal(c) + Decimal(6*k) ) / (Decimal(4.0) * Decimal(c)**Decimal(3.0) * Decimal(mu)**Decimal(2.0))
    B = Decimal(3*k*c) / Decimal(2*mu)
    eps = Decimal(alpha) * Decimal(k)**Decimal(2.0) / (2*mu) + min(A, B)
    return eps

def rdp_sk_sampling(q, alpha, lambda_, c):
    current_sum = Decimal(1-q)**Decimal(alpha-1) * Decimal(alpha*q-q+1)

    j = 2
    current_term = Decimal(comb(alpha, j, exact=True)) * \
                    Decimal(1-q)**Decimal(alpha-j) * Decimal(q)**Decimal(j) * \
                    Decimal(np.e)**Decimal((j-1) * rdp_sk(j, lambda_, c))
    current_sum += current_term
        
    for j in range(3,alpha+1):
        current_term = Decimal(comb(alpha, j, exact=True)) * \
                        Decimal(1-q)**Decimal(alpha-j) * Decimal(q)**Decimal(j) * \
                        Decimal(np.e)**Decimal((j-1) * rdp_sk(j, lambda_, c))
        current_sum += current_term
    return float(current_sum.ln() / Decimal(alpha-1))
```

```python
def naive_group_rdp_sk(alpha, lambda_, q, m):
    """
    Suppose m=2^c for c \in Z
    """
    alpha = alpha * m

    rdp = rdp_sk_sampling(q, alpha, lambda_, 1)
    
    c = np.log2(m)
    
    return float(3**c * rdp)


def exact_rdp_upd_sk(alpha, lambda_, q, m):
    sum_ = Decimal(0.0)
    
    p_r = [Decimal(comb(m, k, exact=True)) * Decimal(q)**Decimal(k) * Decimal(1-q)**Decimal(m-k) for k in range(0, m+1)]
    
    for i in range(0, m+1):
        sum_ +=  p_r[i]*Decimal(np.e)**Decimal( (alpha-1) * rdp_sk_group(alpha, lambda_, 1, i) )
    rdp = sum_.ln() / Decimal(alpha-1)
    return float(rdp)
```

```python
"""
Lower bound of RDP for Skellam
"""
def compute_lbd_sk(q, alpha, lambda_, m):
    sum_ = Decimal(0.0)
    
    p_r = [Decimal(comb(m, k, exact=True)) * Decimal(q)**Decimal(k) * Decimal(1-q)**Decimal(m-k) for k in range(0, m+1)]
    
    for i in range(-10*np.int(np.sqrt(2*lambda_)), 10*np.int(np.sqrt(2*lambda_))):
        sum_up = Decimal(0.0)
        if get_sk_prob(lambda_, i) == Decimal('0'):
            continue

        for k in range(0, m+1):
            sum_up += p_r[k]*get_sk_prob(lambda_, i-k)
        tmp = (Decimal(sum_up) / get_sk_prob(lambda_, i)) **Decimal(alpha)
        tmp = tmp * get_sk_prob(lambda_, i)
        sum_ +=  tmp
    rdp = sum_.ln() / Decimal(alpha-1)
    return float(rdp)
```

```python
def calibrating_sk_noise_lbd(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 5000
    sigma_min = 1
    
    def binary_search(left, right):
        mid = (left + right) / 2
        
        lbd = t*compute_lbd_sk(q, alpha, mid, m)
        ubd = t*compute_lbd_sk(q, alpha, left, m)
        
        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*compute_lbd_sk(alpha, sigma_max, q, m) > tau:
        sigma_max *= 10
        if sigma_max > 1e10:
            print("sigma_max>", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return sigma_max

def calibrating_sk_noise_naive(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 5000
    sigma_min = 1
    
    def binary_search(left, right):
        mid = (left + right) / 2

        lbd = t*naive_group_rdp_sk(alpha, mid, q, m)
        ubd = t*naive_group_rdp_sk(alpha, left, q, m)
        
        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*naive_group_rdp_sk(alpha, sigma_max, q, m) > tau:
        sigma_max *= 10
        if sigma_max > 1e10:
            print("sigma_max >", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return sigma_max


def calibrating_sk_noise_ours(q, m, alpha=4, tau=0.5, t=1, err=1e-4):
    """
    Calibrate noise to privacy budgets
    """
    sigma_max = 5000
    sigma_min = 1
    
    def binary_search(left, right):
        mid = (left + right) / 2

        lbd = t*exact_rdp_upd_sk(alpha, mid, q, m)
        ubd = t*exact_rdp_upd_sk(alpha, left, q, m)
        
        if ubd > tau and lbd > tau:    # min noise & mid noise are too small
            left = mid
        elif ubd > tau and lbd < tau:  # mid noise is too large
            right = mid
        else:
            print("an error occurs in binary search")
            return -1
        return left, right
        
    # check
    while t*naive_group_rdp_sk(alpha, sigma_max, q, m) > tau:
        sigma_max *= 10
        if sigma_max > 1e10:
            print("sigma_max >", sigma_max)
            return -1
    
    while sigma_max-sigma_min > err:
        sigma_min, sigma_max = binary_search(sigma_min, sigma_max)
    return sigma_max
```

## Subsmapled Skellam Varying $m$

```python
q = 0.03

iters = 500

m_list = [16, 32, 64, 128, 256]

alpha = 4
tau = 1

delta = 1e-5

naive_list = []
upd_list = []
lbd_list = []

for m in m_list:
    naive = 1e100
    if abs(np.log2(m)-int(np.log2(m))) < 0.00001:
        naive = calibrating_sk_noise_naive(q, m, alpha=alpha, tau=tau, t=iters, err=0.1)
    upd = calibrating_sk_noise_ours(q, m, alpha=alpha, tau=tau, t=iters, err=0.1)
    lbd = calibrating_sk_noise_lbd(q, m, alpha=alpha, tau=tau, t=iters, err=0.1)
    print("m={}, naive={},upd={},lbd={}".format(m, naive, upd, lbd))
    print("")
    naive_list.append(naive)
    upd_list.append(upd)
    lbd_list.append(lbd)
```
